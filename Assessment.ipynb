```python
import pandas
import matplotlib.pyplot as plt
import numpy as num

df_a = pandas.read_csv('device-a.csv')
df_b = pandas.read_csv('device-b.csv')
```

```python
df_a['cpu0/scaling_cur_freq'].plot(figsize = (15, 10),style='ro',label='cpu0/scaling_cur_freq a')
df_b['cpu0/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu0/scaling_cur_freq b')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Core Frequency [kHz?]')

plt.show()
```

```python
df_a['cpu1/scaling_cur_freq'].plot(figsize = (15, 10),style='ro',label='cpu1/scaling_cur_freq a')
df_b['cpu1/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu1/scaling_cur_freq b')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Core Frequency [kHz?]')

plt.show()
```

```python

plt.subplot(121)
df_b['cpu2/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu2/scaling_cur_freq')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Core Frequency [kHz?]')
plt.subplot(122)
df_b['cpu3/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu3/scaling_cur_freq')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Core Frequency [kHz?]')

plt.show()
```

```python
df_a['BAT0/capacity'].plot(figsize = (15, 10),style='r')
df_b['sbs-1-000b/capacity'].plot(figsize = (15, 10),style='b')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Battery Capacity [%]')

plt.show()
```

```python
meanv_a = int(num.mean(df_a['BAT0/voltage_now']))
meanv_b = int(num.mean(num.absolute(df_b['sbs-1-000b/voltage_now'])))

df_a['BAT0/voltage_now'].plot(figsize = (15, 10),style='r')
df_b['sbs-1-000b/voltage_now'].plot(figsize = (15, 10),style='b')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Battery Voltage [?]')

plt.show()
print("Mean value for the BAT0 battery voltage is " + str(meanv_a) + " and the mean value for the sbs-1-000b battery voltage is " + str(meanv_b))
```

```python
meanc_a = int(num.mean(df_a['BAT0/current_now']))
meanc_b = int(num.mean(num.absolute(df_b['sbs-1-000b/current_now'])))

df_a['BAT0/current_now'].plot(figsize = (15, 10),style='r-')
df_b['sbs-1-000b/current_now'] = num.absolute(df_b['sbs-1-000b/current_now'])
df_b['sbs-1-000b/current_now'].plot(figsize = (15, 10),style='-')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Battery drawn current [?]')

plt.show()

print("Mean value for the BAT0 battery current is " + str(meanc_a) + " and the mean value for the sbs-1-000b battery current is " + str(meanc_b))

power_a = meanv_a*meanc_a
power_b = meanv_b*meanc_b

print("Mean power pulled from the battery of device_b is " + str(power_b/power_a) + " of the power from device_a")
```

```python
df_a['INT3400 Thermal'].plot(figsize = (15, 10),style='r')
df_b['cpu_thermal'].plot(figsize = (15, 10),style='b')
plt.legend()
plt.xlabel('Sample')
plt.ylabel('Temperature [?]')

plt.show()
```
