```python
import pandas
import matplotlib.pyplot as plt

df_a = pandas.read_csv('device-a.csv')
df_b = pandas.read_csv('device-b.csv')
```

```python
df_a['cpu0/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu0/scaling_cur_freq a')
df_b['cpu0/scaling_cur_freq'].plot(figsize = (15, 10),style='ro',label='cpu0/scaling_cur_freq b')
plt.legend()
plt.xlabel('Sample')

plt.show()
```

```python
df_a['cpu1/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu1/scaling_cur_freq a')
df_b['cpu1/scaling_cur_freq'].plot(figsize = (15, 10),style='ro',label='cpu1/scaling_cur_freq b')
plt.legend()
plt.xlabel('Sample')

plt.show()
```

```python

plt.subplot(121)
df_b['cpu2/scaling_cur_freq'].plot(figsize = (15, 10),style='bo',label='cpu2/scaling_cur_freq')
plt.legend()
plt.subplot(122)
df_b['cpu3/scaling_cur_freq'].plot(figsize = (15, 10),style='ro',label='cpu3/scaling_cur_freq')
plt.legend()

plt.show()
```

```python
df_a['BAT0/capacity'].plot(figsize = (15, 10),style='r')
df_b['sbs-1-000b/capacity'].plot(figsize = (15, 10),style='b')
plt.legend()

plt.show()
```

```python
df_a['BAT0/voltage_now'].plot(figsize = (15, 10),style='r')
df_b['sbs-1-000b/voltage_now'].plot(figsize = (15, 10),style='b')
plt.legend()

plt.show()
```

```python
df_a['BAT0/current_now'].plot(figsize = (15, 10),style='r-')
df_b['sbs-1-000b/current_now'] = -df_b['sbs-1-000b/current_now']
df_b['sbs-1-000b/current_now'].plot(figsize = (15, 10),style='-')
plt.legend()

plt.show()
```
